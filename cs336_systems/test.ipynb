```python
warmup_study = []
for i in range(5):
    times = []
    for j in range(10):
        res = benchmarking(
            num_layers=12,
            num_heads=12,
            d_model=768,
            d_ff=3072,
            vocab_size=10000,
            seq_length=256,
            batch_size=4,
            warmup_steps=i,
            timed_steps=10,
            backward=True,
        )
        times.append(res["avg_time_per_step"])
    warmup_study.append(times)
print(warmup_study)
```

```python
import pandas as pd
pd.DataFrame(warmup_study).plot(kind='bar', legend=False)
```
